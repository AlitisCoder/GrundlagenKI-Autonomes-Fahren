```python
#Aufgabe3
wb = erstelle_wissensbasis() # Hier wird die Wissensbasis erstellt, die alle Regeln und Fakten enthält, die für die Inferenz benötigt werden.

# Apply Modus Ponens

# Szenario 1 - Fußgänger vor dem Auto -> Hupe betätigen und anhalten (F = wahr -> H = wahr, A = wahr)

print("=" * 50) # kosmetische Gestaltung, damit es übersichtlich bleibt
print("Szenario 1: Fußgänger vor dem Auto (Fußgänger -> Hupe und Anhalten)")
print("=" * 50)
bekannte_fakten = {AtomareFormel("F")} # Die bekannten Fakten beschreiben einfach nur die Situation
inferenz = wende_modus_ponens_an(wb, bekannte_fakten) # Hier wird die Inferenzmaschine aufgerufen, um die Schlussfolgerungen zu ziehen
print("Abgeleitete Fakten:", inferenz) # Ausgabe der Schlussfolgerungen
print() # Zeilenumbruch für bessere Lesbarkeit
#-------------------------------------------------------------------------

# Szenario 2 - Rote Ampel (Rote Ampel -> Anhalten) (R = wahr -> A = wahr)

print("=" * 50)
print("Szenario 2: Rote Ampel (Rote Ampel -> Anhalten)")
print("=" * 50)
bekannte_fakten = {AtomareFormel("R")}
inferenz = wende_modus_ponens_an(wb, bekannte_fakten)
print("Abgeleitete Fakten:", inferenz)
print()
#-------------------------------------------------------------------------

# Szenario 3 - Freie Fahrt (Kein Fussgänger und keine rote Ampel -> Freie Fahrt) (¬F und ¬R -> ¬H und ¬A)

print("=" * 50)
print("Szenario 3: Freie Fahrt (Kein Fußgänger und keine rote Ampel -> Freie Fahrt)")
print("=" * 50)
bekannte_fakten = {Negation(AtomareFormel("F")), Negation(AtomareFormel("R"))}
inferenz = wende_modus_ponens_an(wb, bekannte_fakten)
print("Abgeleitete Fakten:", inferenz)
print()
#-------------------------------------------------------------------------

```
